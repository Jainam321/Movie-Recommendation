```python
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
movies
```

```python
ratings = pd.read_csv("ratings.csv")
ratings
```

```python
movies_with_ratings = ratings.merge(movies, on='movieId', how='inner')
movies_with_ratings
```

```python
movies_with_ratings2 = movies_with_ratings.drop(["timestamp"],axis=1)
movies_with_ratings2
```

```python
final_df = movies_with_ratings2.sort_values(by=['userId'], ascending=True)
final_df.to_csv('merged_data.csv')
# final_df
```
